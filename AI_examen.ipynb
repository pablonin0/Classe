```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['horsepower'].isnull().sum()
```

```python
mediana_hp = df['horsepower'].median()

df['horsepower'] = df['horsepower'].fillna(mediana_hp)
```

```python
df.info()
```

```python
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']
```

```python
df.head()
```

```python
sea.histplot(df['acceleration'], color = 'blue', bins=20)
```

```python
sea.scatterplot(x=df['displacement'],y= df['horsepower'], color = 'blue')
```

```python
df.head()
```

```python
df_rend = df.groupby('origin')['mpg'].mean()
```

```python
df_rend.head()
```

```python
sea.barplot(df_rend, color='blue')
```
